```python
import pandas as pd 
import seaborn as sns 
```

```python
tips = pd.read_csv('tips.csv')
```

```python
tips.head()
```

```python
titanic = pd.read_csv('Titanic-Dataset.csv')
```

```python
flights = pd.read_csv('flights.csv')
```

```python
flights.head()
```

```python
iris = pd.read_csv('iris.csv')
iris.head()
```

1.Scatterplot (Numerical-Numerical)

```python
sns.scatterplot(x= 'total_bill',y = 'tip',data = tips)
```

```python
sns.scatterplot(x= 'total_bill',y = 'tip',data = tips, hue = 'sex')
```

```python
sns.scatterplot(x= 'total_bill',y = 'tip',data = tips, hue = 'sex', style = 'smoker')
```

```python
sns.scatterplot(x= 'total_bill',y = 'tip',data = tips, hue = 'sex', style = 'smoker',size= 'size')
```

2.Barplot(Numerical - Categorical)

```python
titanic.head()
```

```python
sns.barplot(x='Pclass',y = 'Age', data = titanic)
```

```python
sns.barplot(x='Pclass',y = 'Fare', data = titanic)
```

```python
sns.barplot(x='Pclass',y = 'Fare', data = titanic , hue = 'Sex')
```

```python
sns.barplot(x='Pclass',y = 'Age', data = titanic , hue = 'Sex')
```

3. Box Plot (Numerical - Categorical)

```python
sns.boxplot(x= 'Sex' ,y='Age',data = titanic)
```

```python
sns.boxplot(x= 'Sex' ,y='Age',data = titanic , hue = 'Survived')
```

4. Dist Plot (Numerical - Categorical)

```python
sns.distplot(titanic[titanic['Survived'] == 0]['Age'] )
sns.distplot(titanic[titanic['Survived'] == 1]['Age'] )
```

5. HeatMap( Categorical - Categorical )

```python
pd.crosstab(titanic['Pclass'],titanic['Survived'])
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

```python
titanic.groupby('Pclass')['Survived'].mean()*100
```

```python
(titanic.groupby('Pclass')['Survived'].mean()*100).plot(kind='bar')
```

```python
titanic.groupby('Embarked')['Survived'].mean()*100
```

6.Cluster Map(Categorical - Categorical)

```python
pd.crosstab(titanic['Parch'],titanic['Survived'])
```

```python
sns.clustermap(pd.crosstab(titanic['Parch'],titanic['Survived']))
```

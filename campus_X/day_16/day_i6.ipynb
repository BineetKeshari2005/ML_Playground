converting url json into pd dataframe 

```python
import requests
import pandas as pd

url = 'https://api.exchangerate-api.com/v4/latest/INR'

response = requests.get(url)
data = response.json()

df = pd.DataFrame(data['rates'].items(), columns=['currency', 'rate'])
print(df.head())
```
